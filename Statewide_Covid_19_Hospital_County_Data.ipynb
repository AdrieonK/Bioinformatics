```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# View Dataset
df = pd.read_csv('covid19hospitalbycounty.csv')

# View specific amount of dataset (first 20/86k); (.tail() for end of dataset)
first_counties = df.head(100).tail(100)

# 'X' (counties) selected based on values from 'Y' (patients) in dataset
top_counties = first_counties.sort_values(by='hospitalized_covid_confirmed_patients', ascending=False).head(35)

# Extract county names & hospitalized Covid patients counts from identified top counties
counties = top_counties['county']
patients = top_counties['hospitalized_covid_confirmed_patients']

# Creates an equally spaced array of X-positions
x = np.arange(len(counties))

# Assigns labels to X-positions
plt.bar(x, patients, align='center', alpha=0.5)

# Customize graph

plt.xticks(x, counties, rotation='vertical')
plt.ylabel('Hospitalized & COVID Confirmed')
plt.title('CA Counties with Most Hospitalized Covid Patients')
plt.show()
```
