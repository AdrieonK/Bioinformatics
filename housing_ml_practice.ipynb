```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

```python
melb_data = pd.read_csv('/content/melb_data.csv')

melb_data.head()
```

```python
melb_data = melb_data.dropna()
```

```python
melb_data.head()
```

```python
print(melb_data.dtypes)
```

```python
melb_data.columns
```

```python
scaler = StandardScaler()

data_features = ['Rooms','Distance','Postcode','Bathroom','Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude']

melb_data[data_features] = scaler.fit_transform(melb_data[data_features])
```

```python
y = melb_data.Price
```

```python
melb_features = ['Rooms','Postcode','Bathroom','Landsize','YearBuilt','Lattitude','Longtitude']

X = melb_data[melb_features]

X.head()
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y)
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
  model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 1)
  model.fit(train_X, train_y)
  val_preds = model.predict(val_X)
  mae = mean_absolute_error(val_y, val_preds)
  return mae
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
```

```python
scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candidate_max_leaf_nodes}

best_tree_size = min(scores, key = scores.get)

```

```python
final_model = DecisionTreeRegressor(max_leaf_nodes = best_tree_size, random_state = 1)
final_model.fit(X, y)
decision_tree_predictions = final_model.predict(X)
decision_tree_mae = mean_absolute_error(y, decision_tree_predictions)

print("Decision Tree MAE: ", decision_tree_mae)
```

```python
rf_model = RandomForestRegressor(random_state = 1)

rf_model.fit(train_X, train_y)

rf_model_predictions = rf_model.predict(val_X)
rf_model_mae = mean_absolute_error(val_y, rf_model_predictions)

print(rf_model_mae)
```
